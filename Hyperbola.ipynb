```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi, sqrt
```

```python
class Hyperbola:
    def __init__(self, center, distance, cjg_axis):
        self.center = center
        self.distance = distance
        self.cjg_axis = cjg_axis
        self.hyperbola = self._hyperbola()
    
    def _hyperbola(self):
        a = self.distance
        b = self.cjg_axis
        
        return (a**2 + b**2)**0.5
    
    def get_specs(self, orientation):
        h, k = self.center
        a = self.distance
        b = self.cjg_axis
        c = self.hyperbola
    
        if orientation == "horizontal":
            vertices = (f"({h + c}, {k}) and ({h - a}, {k})")
            foci = (f"({h + c}, {k}) and ({h - c}, {k})")
            asymptotes = (f"y = {k} + ({b}/{a})(x - {h}) and y = {k} - ({b}/{a})(x - {h})")
            transverse_axis = (f"x = {k}")
            conjugate_axis = (f"y = {h}")
    
            specs = vertices, foci, asymptotes, transverse_axis, conjugate_axis
        
        elif orientation == "vertical":
            vertices = (f"({h}, {k + a}) and ({h}, {k - a})")
            foci = (f"({h}, {k + c}) and ({h}, {k - c})")
            asymptotes = (f"y = {k} + ({a}/{b})(x - {h}) and y = {k} - ({a}/{b})(x - {h})")
            transverse_axis = (f"y = {h}")
            conjugate_axis = (f"x = {k}")
    
            specs = vertices, foci, asymptotes, transverse_axis, conjugate_axis
        
        else:
            specs = 'error'
    
        return specs

    def plot(self, orientation):
        """
        This method uses the plot_hyperbola function to visualize the hyperbola.
        """
        h, k = self.center
        a = self.distance
        b = self.cjg_axis

        # Call the plotting function
        plot_hyperbola(center=(h, k), a=a, b=b, orientation=orientation)

def plot_hyperbola(center, a, b, orientation):
"""
    Plots the hyperbola along with its key elements: vertices, foci, asymptotes,
    transverse and conjugate axes based on orientation.
    
    Parameters:
        center (tuple): The center of the hyperbola (h, k).
        a (float): The length of the transverse axis (semi-major axis).
        b (float): The length of the conjugate axis (semi-minor axis).
        orientation (str): Orientation of the hyperbola ('horizontal' or 'vertical').
    """
    h, k = center
    c = (a**2 + b**2)**0.5  # Foci distance

    # Create a range of x values for plotting
    x_vals = np.linspace(h - 2*a, h + 2*a, 400)
    
    # Plot based on orientation
    if orientation == 'horizontal':
        # Hyperbola curves (horizontal)
        y_vals_positive = k + (b/a) * np.sqrt(x_vals - h)
        y_vals_negative = k - (b/a) * np.sqrt(x_vals - h)

        # Asymptotes (lines with slope ±b/a through the center)
        asymptotes_positive = k + (b/a)*(x_vals - h)
        asymptotes_negative = k - (b/a)*(x_vals - h)
        
        vertices = [(h + a, k), (h - a, k)]
        foci = [(h + c, k), (h - c, k)]
        
    elif orientation == 'vertical':
        # Hyperbola curves (vertical)
        y_vals_positive = k + (a/b) * np.sqrt(x_vals - h)
        y_vals_negative = k - (a/b) * np.sqrt(x_vals - h)

        # Asymptotes (lines with slope ±a/b through the center)
        asymptotes_positive = k + (a/b)*(x_vals - h)
        asymptotes_negative = k - (a/b)*(x_vals - h)

        vertices = [(h, k + a), (h, k - a)]
        foci = [(h, k + c), (h, k - c)]

    # Plot setup
    plt.figure(figsize=(8, 6))

    # Plot asymptotes
    plt.plot(x_vals, asymptotes_positive, 'r--', label="Asymptote 1")
    plt.plot(x_vals, asymptotes_negative, 'r--', label="Asymptote 2")

    # Plot hyperbolic curves
    plt.plot(x_vals, y_vals_positive, 'b-', label="Hyperbola Curve (positive)")
    plt.plot(x_vals, y_vals_negative, 'b-', label="Hyperbola Curve (negative)")

    # Plot vertices
    for vertex in vertices:
        plt.plot(vertex[0], vertex[1], 'bo', label="Vertex" if vertex == vertices[0] else "")

    # Plot foci
    for focus in foci:
        plt.plot(focus[0], focus[1], 'go', label="Focus" if focus == foci[0] else "")

    # Plot transverse axis (vertical or horizontal)
    if orientation == 'horizontal':
        plt.axvline(x=h, color='g', linestyle='-', label="Transverse Axis")
    elif orientation == 'vertical':
        plt.axhline(y=k, color='b', linestyle='-', label="Transverse Axis")

    # Plot conjugate axis (perpendicular to the transverse axis)
    if orientation == 'horizontal':
        plt.axhline(y=k, color='b', linestyle='-', label="Conjugate Axis")
    elif orientation == 'vertical':
        plt.axvline(x=h, color='g', linestyle='-', label="Conjugate Axis")

    # Set plot limits
    plt.xlim(h - 2*a, h + 2*a)
    plt.ylim(k - 2*c, k + 2*c)

    # Labels and legend
    plt.title(f'Hyperbola with {orientation.capitalize()} Orientation')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.legend()

    # Show the plot
    plt.show()
```

```python
def get_nonzero_input(var_name):
    while True:
        value = eval(input(f"{var_name}: "))  # console
        if value == 0:
            print(f"Invalid input! '{var_name}' cannot be zero. Try again.")
        else:
            return value

def print_specs(vertices, foci, asymptotes, transverse_axis, conjugate_axis):
    print(f"Vertices: {vertices}")
    print(f"Foci: {foci}")
    print(f"Asymptotes: {asymptotes}")
    print(f"Transverse Axis: {transverse_axis}")
    print(f"Conjugate Axis: {conjugate_axis}")
```

```python
#### print(f"You selected: Hyperbola")
print("\nEnter the values for h, k, a, and b:")

h = eval(input("h: ")) #console
k = eval(input("k: ")) #console

a = get_nonzero_input("a") #console
b = get_nonzero_input("b") #console        

print(f"\nThe equation of the Hyperbola is: (x-{k})^2 / {a**2}) - ((y-{k})^2 / {b**2}) = 1")

hyperbola = Hyperbola(center=(h, k), distance=a, cjg_axis=b)
```

```python
print(f"\nChoose the orientation of the Hyperbola: ")
orientation = input("Type 'horizontal' or 'vertical': ")

if orientation == "horizontal":
    output = hyperbola.get_specs('horizontal')
    vertices, foci, asymptotes, transverse_axis, conjugate_axis = output

    print_specs(vertices, foci, asymptotes, transverse_axis, conjugate_axis)
    hyperbola.plot(orientation)
    
elif orientation == "vertical":
    output = hyperbola.get_specs('vertical')
    vertices, foci, asymptotes, transverse_axis, conjugate_axis = output

    print_specs(vertices, foci, asymptotes, transverse_axis, conjugate_axis)
    hyperbola.plot(orientation)

else:
    print(f"Invalid orientation choice. ")
    print(f"With center at ({h}, {k}), transverse axis length {2*a}, and conjugate axis length {2*b}")
```
